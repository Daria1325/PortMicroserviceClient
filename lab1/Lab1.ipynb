2. NumPy (створити набір даних (розмір за власним рішенням), провести CRUD операції над даними (скаляр, вектор, матриця)). Властивість shape та метод reshape. Операції над масивами. Використати arange, random.

```python
import numpy as np

#створити набір даних
A = np.array([1,5,4,7,8,9,6,4,5])
print("A = ",A)

#провести CRUD операції
#скаляр
a = np.random.randint(10,20)
print("\nscalar\n", a)
a = 5
print("update\n",a)

#вектор
b = np.arange(3, 10, 1)
print("\nvector\n", b)
np.random.shuffle(b)
print("update\n",b)
b = np.append(b,[1,2,3],0)
print("update\n",b)

#матриця
c = np.array([[4,7,8],[5,7,6],[10,7,8]])
print("\nmatrix\n",c)
c[0][0] = 5
print("update\n",c)
#видалимо 1 стовбець оригіальний масив не змінюється (повертається копія)
print("delete\n",np.delete(c, 0, 1))
#видалимо 2 рядок
print("delete\n",np.delete(c, 1, 0))



print("shape A \n",A.shape)
A = A.reshape((3,3))
print("reshape A \n",A)
```

```python
#операції над масивами
print("A\n", A)
print("c\n",c)
print("A+c\n",A+c)

print("A*c\n",A*c)
print("dot(A,c)\n",np.dot(A,c))
print("A*5\n",A*5)
print("A-c\n",A-c)
print("A/c\n",A/c)
```

3. Pandas (створити dataframe різними способами, провести маніпуляції з даними (head, describe, iloc, loc, індексація з умовами відбору,групування, пошук «пустих» значень, сортування.);

```python
import pandas as pd

#Створення dataframe з файлу
data1 = pd.read_csv('color_srgb.csv')
print("\nСтворення dataframe з файлу\n")
display(data1)

#Створення dataframe з даних
d = {'Name': ['Sasha', 'Luca', 'Berta','Sasha'], 'Age': [30, 45,10,20],'Sex': ['Male', 'Male','Female','Male'],'Education': ['Bachalor', 'Bachalor', None,None]}
data2 = pd.DataFrame(data=d)
print("\nСтворення dataframe з даних\n")
display(data2)

#head
print("\nhead\n")
display(data1.head(4))

#describe
print("\ndescribe\n",)
print(data1.describe())
print(data2.describe())
#iloc
print("\niloc\n")
display(data1.iloc[:5, :2])
#loc
print("\nloc\n")
display(data1.loc[2:7,'HEX'])

#індексація з умовами відбору
condition1 = data2["Age"] >= 25
condition2 = data2["Sex"] == 'Male'
data2_subset = data2[(condition1&condition2)]
display(data2_subset)

#групування
print("\ngroup by name\n")
df_count = data2.groupby(['Name']).count()
display(df_count)

#пошук "пустих" значень
print("\nfind nulls\n")
display(data2.isnull().sum())

#сортування
print("\nsort by age\n")
df_sort = data2.sort_values(['Age'])
display(df_sort)
```

4. Matplotlib (створити набір даних лінійної функції з урахуванням помилок вимірювань, побудувати графік функції та згенерованих даних, обчислити похибку з використанням метрик MAE, MSE). Записати отримані результати у csv файл (формат: X,Y, Y_hat, mAE, mSE)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error


kol = 50

data = np.ones( (5,kol)) 

def func (x):
  return 5*x

def funcWithError (x):
  np.random.seed(int(x*10))
  t = np.random.randint(-2,2)/1.5
  if(np.random.randint(0,1)%2):
    return 5*x-t
  else:
    return 5*x+t

for i in range(kol):
  np.random.seed(i)
  data[0][i] = np.random.randint(0,30)/10
  data[1][i] = func(data[0][i])
  data[2][i] = funcWithError(data[0][i])
  data[3][i] = mean_absolute_error([data[2][i]],[data[1][i]])
  data[4][i] = mean_squared_error([data[2][i]],[data[1][i]])

dataTranspose = data.T
data_to_csv = pd.DataFrame(dataTranspose,columns=['X','Y','Y_hat','mAE','mSE'])
data_to_csv.to_csv('lab1_data_task3.csv', index=False)


print("MAE",mean_absolute_error(data[2], data[1]))
print("MSE",mean_squared_error(data[2], data[1]))

fig = plt.figure(figsize=(10, 6))

axes = fig.add_axes([0,0,1,1])

axes.plot(data[0], data[1], 'r')
axes.scatter(data[0],data[2],c='b')
axes.legend([r'$5x$'], loc=0)

plt.show()
```
